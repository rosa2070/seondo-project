# 여기부터 진짜 시작
전처리 과정 하나의 함수로 만들기

```python
def preprocessing(review, okt, remove_stopwords = False, stop_words = []):
    # 함수의 인자는 다음과 같다.
    # review : 전처리할 텍스트
    # okt : okt 객체를 반복적으로 생성하지 않고 미리 생성후 인자로 받는다.
    # remove_stopword : 불용어를 제거할지 선택 기본값은 False
    # stop_word : 불용어 사전은 사용자가 직접 입력해야함 기본값은 비어있는 리스트

    # 1. 한글 및 공백을 제외한 문자 모두 제거.
    content_text = BeautifulSoup(content, "html5lib").get_text() # HTML 태그 제거
    content_text = re.sub("\n", " ", content_text)
    content_text = re.sub("[^ㄱ-ㅎㅏ-ㅣ가-힣a-zA-Z]", " ", content_text)

    # 2. okt 객체를 활용해서 형태소 단위로 나눈다.
    word_context = okt.morphs(content_text, stem=True)

    if remove_stopwords:

        # 불용어 제거(선택적)
        word_context = [token for token in word_context if not token in stop_words]


    return word_context
```

```python
stop_words = [ '은', '는', '이', '가', '하', '아', '것', '들','의', '있', '되', '수', '보', '주', '등', '한']
okt = Okt()
clean_train_review = []

for content in train_data['content']:
    # 비어있는 데이터에서 멈추지 않도록 string인 경우만 진행
    if type(content) == str:
        clean_train_review.append(preprocessing(content, okt, remove_stopwords = True, stop_words=stop_words))
    else:
        clean_train_review.append([])  #string이 아니면 비어있는 값 추가

```

# 데이터 중 결측치 제거
특수문자, 숫자로만 구성된 단어, 1글자 단어제거

```python
test = []
for doc in train_data["clean_word"]:
    for word in doc:
    	# 특수문자, 숫자로만 구성된 단어, 1글자 단어
        if word.isalnum() == False or word.isdigit() == True or len(word) == 1:
#             print("특수")
            continue
        test.append(word)

# 등장 단어리스트 추출
vocab = list(set(test))
vocab.sort()
```

```python
vocab
```

# tf값 계산
각 문서별 등장 단어에 대한 tf값이 정리된 테이블 저장

```python
def tf(t, d):
    return d.count(t)

N = len(train_data)

result = []
for i in range(N): # 각 문서에 대해서 아래 명령을 수행
    result.append([])
    d = train_data["content"][i]

    for j in range(len(vocab)):
        t = vocab[j]
        result[-1].append(tf(t, d))

tf_ = pd.DataFrame(result, columns = vocab)
```

```python
tf_
```

```python
tf_.sum().sort_values(ascending = False)[:50]
```

# idf값 계산

```python
from math import log


def idf(t):
    df = 0
    for doc in train_data["clean_word"]:
        df += t in doc
    return log(N/(df + 1))


result = []
for j in range(len(vocab)):
    t = vocab[j]
    result.append(idf(t))


idf_ = pd.DataFrame(result, index = vocab, columns = ["IDF"])
```

```python
idf_.sort_values(by = 'IDF', ascending= False)
```

# tf-idf 값 계산

```python
def tfidf(t, d):
    return tf(t,d)* idf(t)

result = []
for i in range(N):
    result.append([])
    d = train_data["clean_word"][i]
    for j in range(len(vocab)):
        t = vocab[j]

        result[-1].append(tfidf(t,d))
```

```python
tfidf_ = pd.DataFrame(result, columns = vocab)
```
